```python
import serial
import cv2
import numpy as np

# Inisialisasi komunikasi serial
try:
    arduino = serial.Serial('COM3', 115200, timeout=1)
    print("Serial connected ke Arduino.")
except:
    arduino = None
    print("Gagal koneksi ke Arduino. Lanjut tanpa kirim serial.")

# Fungsi kirim perintah
def send_command(action, speed=200):
    command = f"{action},{speed}\n"
    if arduino:
        arduino.write(command.encode())
        print(f"Kirim: {command.strip()}")
    else:
        print(f"[Simulasi] {command.strip()}")

# Buka kamera
cap = cv2.VideoCapture(0)

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            continue

        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        # Rentang merah
        lower_red1 = np.array([0, 100, 100])
        upper_red1 = np.array([10, 255, 255])
        lower_red2 = np.array([160, 100, 100])
        upper_red2 = np.array([180, 255, 255])

        # Masking dua area merah
        mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
        mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
        mask = cv2.bitwise_or(mask1, mask2)

        # Deteksi kontur
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        merah_terdeteksi = False

        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area > 500:
                merah_terdeteksi = True
                break  # cukup satu kontur besar

        if merah_terdeteksi:
            send_command("STOP", 0)
            cv2.putText(frame, "STOP: Merah Besar Terdeteksi", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
        else:
            send_command("MOVE_FORWARD", 200)
            cv2.putText(frame, "MAJU", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

        # Tampilkan
        cv2.imshow("AI Controller", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except KeyboardInterrupt:
    print("Dihentikan manual")

# Bersihkan
cap.release()
cv2.destroyAllWindows()
if arduino:
    arduino.close()
```
